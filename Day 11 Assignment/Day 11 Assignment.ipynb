```python
import pandas as pd
from scipy.stats import pearsonr


dataset=pd.read_csv("general_data.csv")

dataset.head(20)
```

```python
dataset['Attrition_new'] = dataset.Attrition.map({'Yes':1, 'No':0})
```

- since attrition colmn has str variables (categorical valriables) we need to convert them to ints(non cat vars)

```python
dataset.head()
```

```python
dataset
```

- h0 = there is no correlation b/n attrition and age
- h1 = there is correlation

# Correlation b
## Attribute and Age

```python
from scipy.stats import pearsonr
import matplotlib.pyplot as plt

stats,p=pearsonr(dataset.Attrition_new,dataset.Age)

print("Correlation is -",stats,"\nP-value is - ",p)
if p<0.05:
    print("H0 rejected i.e there is no correlation")
else:
    print("H1 is accepted i.e there is correlation")
```

```python
sampledata=dataset.sample(30)
plt.scatter(sampledata.Attrition_new,sampledata.Age)
```

## Attrition and Distance from home¶
- h0 = there is no correlation betn attritoin and distfromhome
- abs h1 = there is correlation

```python
stats,p=pearsonr(dataset.Attrition_new,dataset.DistanceFromHome)
print("Correlation is -",stats,"\nP-value is - ",p)
if p<0.05:
    print("H0 rejected i.e there is no correlation")
else:
    print("H1 is accepted i.e there is correlation")
```

```python
plt.scatter(dataset.Attrition_new,dataset.DistanceFromHome)
```
